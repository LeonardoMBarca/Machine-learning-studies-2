```python
data = "/Users/leona/OneDrive/Documentos/GitHub/Machine-learning-studies-2/ensemble/random_forest/exercise/BRAZIL_CITIES.csv"
```

```python
df = pd.read_csv(data, sep=';', decimal=',')
```

```python
df.info()
```

```python
df.drop(['CITY', 'IDHM Ranking 2010', 'IDHM_Renda', 'IDHM_Longevidade', 'IDHM_Educacao', 'LONG', 'LAT', 'GVA_MAIN', 'REGIAO_TUR', 'MUN_EXPENDIT', 'HOTELS', 'BEDS', 'Pr_Agencies', 'Pu_Agencies', 'Pr_Bank', 'Pu_Bank', 'Pr_Assets', 'Pu_Assets','UBER','MAC','WAL-MART'], axis=1, inplace=True)
```

```python
df['CATEGORIA_TUR'].value_counts()
```

```python
df.isna().sum() / len(df) * 100
```

```python
for i in df.columns:
    percent = df[i].isna().sum()/len(df)*100
    if percent < 40:
        df.dropna(subset=i, inplace=True)
```

```python
df.shape
```

```python
encode_state = pd.get_dummies(df['STATE'])
encode_state = encode_state.astype(int)
df = pd.concat([df, encode_state], axis=1)
```

```python
encode_ru = pd.get_dummies(df['RURAL_URBAN'])
encode_ru = encode_ru.astype(int)
df = pd.concat([df, encode_ru], axis=1)
```

```python
df.drop(columns=['RURAL_URBAN', 'STATE'], axis=1, inplace=True)
```

```python
df.head()
```

```python
x_train = df[df['CATEGORIA_TUR'].notnull()]
x_train.drop('CATEGORIA_TUR', axis=1, inplace=True)
y_train = df[df['CATEGORIA_TUR'].notnull()]['CATEGORIA_TUR']
x_test = df[df['CATEGORIA_TUR'].isnull()]
x_test.drop('CATEGORIA_TUR', axis=1, inplace=True)
y_test = df[df['CATEGORIA_TUR'].isnull()]['CATEGORIA_TUR']
```

```python
for i in x_train.columns:
    if x_train[i].dtype == object:
        print(i)
        x_train[i] = x_train[i].str.replace(',', '')
        x_train[i] = x_train[i].astype(float)
for i in x_test.columns:
    if x_test[i].dtype == object:
        print(i)
        x_test[i] = x_test[i].str.replace(',', '')
        x_test[i] = x_test[i].astype(float)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)

result = model.predict(x_test)
df.CATEGORIA_TUR[df['CATEGORIA_TUR'].isnull()] = result
```

```python
encode_tur = pd.get_dummies(df['CATEGORIA_TUR'])
encode_tur = encode_tur.astype(int)
df = pd.concat([df, encode_tur], axis=1)
df.drop(columns=['CATEGORIA_TUR'], axis=1, inplace=True)
```

```python
for i in df.columns:
    if df[i].dtype == object:
        print(i)
        df[i] = df[i].str.replace(',', '')
        df[i] = df[i].astype(float)
```

MODEL

```python
x = df.drop('IDHM', axis=1)
y = df['IDHM']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
model = RandomForestRegressor(n_estimators=50)
result = cross_val_score(model, x, y, n_jobs=-1)

print(f'The score of this model was: {result.mean()}')
```
